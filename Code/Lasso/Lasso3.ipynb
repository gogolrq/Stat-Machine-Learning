```python
%reset -f
import numpy as np
from sklearn import linear_model
import torch
n=100
p=20
alpha=0.1
x=np.random.normal(0, 1, (n,p))
xx=np.concatenate((np.ones((n,1)),x), axis=1)
b=np.concatenate(([-2],np.ones(5), np.zeros(p-5)), axis=0)
y=xx @ b+np.random.normal(0, 1, n)
y
lasso=linear_model.Lasso(alpha=alpha)
lasso.fit(X=x,y=y)
print(lasso.coef_)

ols=linear_model.LinearRegression()
ols.fit(X=x, y=y)
print(ols.coef_)
#np.vstack((lasso.coef_, ols.coef_))
```

```python
xx_tensor= torch.from_numpy(xx)
y_tensor= torch.from_numpy(y)
print(p)
theta=torch.rand(p+1,requires_grad=True, dtype=torch.float64)
#define loss function


def loss():
    l=torch.pow(y_tensor-xx_tensor @ theta, 2)
    return alpha*torch.abs(theta[1:]).sum()+l.mean()/2
    #return l.mean()/2

gamma = 0.05 # learning rate
n_iters = 2000 # number of iterations


for epoch in range(n_iters):
    # making a pridiction in forward pass
    l = loss()
    # backward pass for computing the gradients of the loss w.r.t to learnable parameters
    l.backward()
    # updateing the parameters after each iteration
    theta.data = theta.data - theta.grad*gamma
    # zeroing gradients after each iteration
    theta.grad.zero_()
    
    print(f'epoch {epoch+1}: loss = {l:.3f}')


```
